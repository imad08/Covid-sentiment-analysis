```python
import pandas as pd
df = pd.read_csv('dhar_clean.csv')
```

```python
df.head()
```

```python
df_dictionary = pd.read_csv('data_dictionary (2).csv')
df_dictionary.head()
```

```python
keywords = []
for word in df_dictionary.making:
  keywords.append(word)
keywords.append('making')
```

```python
keywords
```

```python
df.dates.describe()
```

```python
df['frequency']=0
for word in keywords:
  df['frequency'] += df['tweet'].map(lambda x: x.count(word))
```

```python
df.head()
size=df.shape[0]
```

```python
sum=[]
for i in range(29):
  s=0
  for j in range(size):
    if(df['dates'][j]==i):
      s=s+df['frequency'][j]
  sum.append(s)
```

```python
sum
```

```python
import datetime 
```

```python
start_date = datetime.date(2021, 4, 22)
end_date = datetime.date(2021,5,20)
delta = datetime.timedelta(days=1)
date=[]
while start_date <= end_date:
    print(start_date)
    date.append(start_date)
    start_date += delta
```

```python
final=pd.DataFrame(columns=["date","freq"])
final['date'] = date
final['freq'] = sum
```
