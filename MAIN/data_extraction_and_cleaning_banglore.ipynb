```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import nltk 
import string
import re
%matplotlib inline
pd.set_option('display.max_colwidth', 100)
```

```python
def load_data():
    data = pd.read_csv('combined_banglore.csv')
    return data
```

```python
tweet_df = load_data()
tweet_df.head()
```

```python
tweet_df = tweet_df.drop(['id','conversation_id','created_at','timezone','place','cashtags','user_id_str','username','name','day','hour','link','urls','photos','video','thumbnail','quote_url','search','near','geo','source','user_rt_id','user_rt','retweet_id','reply_to','retweet_date','translate','trans_src','trans_dest','Unnamed: 0'],axis = 1)
```

```python
tweet_df.info()
```

```python
print('Dataset size:',tweet_df.shape)
print('Columns are:',tweet_df.columns)
```

```python
tweet_df.isnull().sum()
```

```python
tweet_df.isnull().sum()
```

```python
tweet_df.dropna(inplace=True)
```

```python
tweet_df.reset_index(drop=True,inplace=True)
```

```python
string.punctuation
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

tweet_df['tweet'] = tweet_df['tweet'].apply(lambda x: remove_punct(x))
tweet_df.head(10)
```

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

tweet_df['Tweet_tokenized'] = tweet_df['tweet'].apply(lambda x: tokenization(x.lower()))
tweet_df.head()
```

```python
import nltk
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
tweet_df['Tweet_nonstop'] = tweet_df['Tweet_tokenized'].apply(lambda x: remove_stopwords(x))
tweet_df.head(10)
```

```python
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

tweet_df['Tweet_stemmed'] = tweet_df['Tweet_nonstop'].apply(lambda x: stemming(x))
tweet_df.head()
```

```python
import nltk
nltk.download('wordnet')
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

tweet_df['Tweet_lemmatized'] = tweet_df['Tweet_nonstop'].apply(lambda x: lemmatizer(x))
tweet_df.head()
```

```python
ntweet_df= tweet_df[tweet_df.language == 'en']
```
